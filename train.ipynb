# Numpy

```python
import numpy as np
np.random.seed(42)
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
import matplotlib.pyplot as plt
import tensorflow as tf
```

## Define a function to convert each element of a column of the dataset into number and then normalize the column

```python
def convert_normalize(df, column, values):
    d = { values[i] : i for i in range(0, len(values) ) }
    # Convert from character to numbers
    dfc = df[column].map(d).astype(float)
    # Then normalize the column, subtracting mean and then dividing by std
    # convert to 0 all NaN (in case of some divide by zero...)
    df[column] = ((dfc-dfc.mean(axis=0))/dfc.std(axis=0)).fillna(0)
```

### Define a function that for each column of a dataset,
### call adjustcolumn() passing the name (the index) and
### a list of possible values retrieved by Attribute Information

```python
def convert_normalize_dataset(df):
    convert_normalize(df, "cap-shape", ["b", "c", "x", "f", "k", "s"])
    convert_normalize(df, "cap-shape", ["b", "c", "x", "f", "k", "s"])
    convert_normalize(df, "cap-surface", ["f", "g", "y", "s"])
    convert_normalize(df, "cap-color", ["n", "b", "c", "g", "r", "p", "u", "e", "w", "y"])
    convert_normalize(df, "bruises", ["t", "f"])
    convert_normalize(df, "odor", ["a", "l", "c", "y", "f", "m", "n", "p", "s"])
    convert_normalize(df, "gill-attachment", ["a", "d", "f", "n"])
    convert_normalize(df, "gill-spacing", ["c", "w", "d"])
    convert_normalize(df, "gill-size", ["b", "n"])
    convert_normalize(df, "gill-color", ["k", "n", "b", "h", "g", "r", "o", "p", "u", "e", "w", "y"])
    convert_normalize(df, "stalk-shape", ["e", "t"])
    convert_normalize(df, "stalk-root", ["b", "c", "u", "e", "z", "r", "?"])
    convert_normalize(df, "stalk-surface-above-ring", ["f", "y", "k", "s"])
    convert_normalize(df, "stalk-surface-below-ring", ["f", "y", "k", "s"])
    convert_normalize(df, "stalk-color-above-ring", ["n", "b", "c", "g", "o", "p", "e", "w", "y"])
    convert_normalize(df, "stalk-color-below-ring", ["n", "b", "c", "g", "o", "p", "e", "w", "y"])
    convert_normalize(df, "veil-type", ["p", "u"])
    convert_normalize(df, "veil-color", ["n", "o", "w", "y"])
    convert_normalize(df, "ring-number", ["n", "o", "t"])
    convert_normalize(df, "ring-type", ["c", "e", "f", "l", "n", "p", "s", "z"])
    convert_normalize(df, "spore-print-color", ["k", "n", "b", "h", "r", "o", "u", "w", "y"])
    convert_normalize(df, "population", ["a", "c", "n", "s", "v", "y"])
    convert_normalize(df, "habitat", ["g", "l", "m", "p", "u", "w", "d"])
```

### Define a function that returns converted target column of a dataset

```python
def convert_target_column(df):
    return pd.get_dummies(pd.Series(data=df["class"])).to_numpy().tolist()
```

### Load dataset from train file and normalize

```python
# Open the train file for reading...
train = pd.read_csv('train.csv')
convert_normalize_dataset(train)
```

### Load dataset from test file and normalize

```python
# Open the test file for reading...
test = pd.read_csv('test.csv')
convert_normalize_dataset(test)
```

### Create X and y train data

```python
y_train = convert_target_column(train)
X_train = train.drop('class', axis=1).to_numpy()
```

### Create X and y test data

```python
y_test = convert_target_column(test)
X_test = test.drop('class', axis=1).to_numpy()
```

```python
# Convert sets because of tensorflow
X_train = tf.convert_to_tensor(X_train, np.float32)
y_train = tf.convert_to_tensor(y_train, np.float32)

# https://www.tensorflow.org/api_docs/python/tf/keras/Sequential
# Sequential groups a linear stack of layers into a tf.keras.Model.
model = tf.keras.Sequential()

# https://www.tensorflow.org/api_docs/python/tf/keras/layers/Dense
# Add regular densely-connected NN layer, with 22 units (input size)
model.add(tf.keras.layers.Dense(22, activation=tf.nn.relu ))
# https://www.tensorflow.org/api_docs/python/tf/keras/layers/Dropout
# Applies Dropout to the input (of this NN layer) with half rate.
model.add(tf.keras.layers.Dropout(0.5))

# Add another regular densely-connected NN layer, with 12 units (nearly half of input size)
model.add(tf.keras.layers.Dense(12, activation=tf.nn.relu ))
# Applies Dropout to the input (of this NN layer) with half rate.
model.add(tf.keras.layers.Dropout(0.5))

#Add last NN layer, with 2 units that represent the two outputs
model.add(tf.keras.layers.Dense(2, activation=tf.nn.relu ))

# Learning rate
lr = 0.1

# Stochastic Gradient Descent as optimizer
optimizer = tf.optimizers.SGD(lr=lr)
# MeanSquaredError as loss function
loss = tf.losses.MeanSquaredError()

# Compile model
model.compile(loss=loss,
              optimizer=optimizer, 
              metrics=['accuracy'])

# fit the model: do the train...
history = model.fit(X_train, y_train, epochs=2000)
plt.plot(history.history['loss'])
```

### Train accuracy:

```python
history.history['accuracy'][-1]
```

### Test accuracy:

```python
results = model.evaluate(
    tf.convert_to_tensor(X_test, np.float32),
    tf.convert_to_tensor(y_test, np.float32), 
    batch_size=128)
print('test acc:', results[1])
```
